# Variação linear da velocidade com a profundidade

O modelo de em que a velocidade cresce linearmente com a profundidade é definido pela função:

$$v(z)=v_0+az$$

Onde $a$ é o gradiente linear de velocidade:

$$a\geq0$$

Para este modelo, no _turning point_,

$$pv(z)=p(v_0+az)=1$$

Então,

$$z_{tp}=\frac{1-pv_0}{pa}$$

Partindo da equação do tempo de trânsito $T$:

$$T=2\int_{0}^{H}\frac{dz}{v(z)\sqrt{1-p^2v^2(z)}}$$

Se,

$$y=p(v_0+az)$$

$$dy=padz$$

$$y(0)=pv_0$$

$$y(H)=pv_0+paH=pv_H$$

A integral para T,

$$T=\frac{2}{a}\int_{0}^{H}\frac{apdz}{pv(z)\sqrt{1-p^2v^2(z)}}=\frac{2}{a}\int_{0}^{H}\frac{dy}{y\sqrt{1-y^2}}$$

O resultado da integral é definido na tabela de integrais. Assim,

$$\int_{0}^{H}\frac{dy}{y\sqrt{1-y^2}}=-\ln\Bigg|\frac{1+\sqrt{1^2-y^2}}{y}\Bigg|$$

Deste modo,

$$T=\frac{-2}{a}\ln\Bigg|\frac{1+\sqrt{1-y^2}}{y}\Bigg|_{y=pv_0}^{y=pv_H}$$

Invertendo os limites,

$$T=\frac{2}{a}\ln\Bigg|\frac{1+\sqrt{1-y^2}}{y}\Bigg|_{y=pv_H}^{y=pv_0}$$

Logo,

$$T=\frac{2}{a}\left\{\ln\Bigg|\frac{1+\sqrt{1-p^2v^2_0}}{pv_0}\Bigg|-\ln\Bigg|\frac{1+\sqrt{1-p^2v^2_H}}{pv_H}\Bigg|\right\}$$

Utilizando a propriedade dos logarítmos,

$$T=\frac{2}{a}\ln\left[\frac{1+\sqrt{1-p^2v^2_0}}{pv_0}\frac{pv_H}{1+\sqrt{1-p^2v^2_H}}\right]$$

Por fim,

$$T=\frac{2}{a}\ln\left[\frac{v_H}{v_0}\frac{1+\sqrt{1-p^2v^2_0}}{1+\sqrt{1-p^2v^2_H}}\right]$$

Para a distância horizontal percorrida pelo raio,

$$X=\int_{0}^{H}\frac{pv(z)}{\sqrt{1-p^2v^2(z)}}dz$$

Corresponde a metade da distância. Se,

$$X=\int_{0}^{H}\frac{pv}{\sqrt{1-p^2v^2}}dz=\frac{1}{ap}\int_{0}^{H}\frac{ap^2v}{\sqrt{1-p^2v^2}}dz
=\frac{1}{ap}\int_{0}^{H}\frac{ap^2(v_0+az)}{\sqrt{1-p^2v^2}}dz$$

Assim,

$$X=-\frac{1}{ap}\sqrt{1-p^2(v_0+az)^2}\big|_{0}^{H}$$

Pois, para,

$$X=\frac{1}{ap}\int_{0}^{H}\frac{ap^2v}{\sqrt{1-p^2v^2}}dz=\frac{1}{ap}\int_{0}^{H}\frac{ap^2(v_0+az)}{\sqrt{1-p^2(v_0+az)^2}}dz$$

Se,

$$u=\sqrt{1-p^2(v_0+az)^2}$$

$$du=\frac{d}{dz}\sqrt{1-p^2(v_0+az)^2}dz$$

E,

$$du=\frac{1}{2}\frac{1}{\sqrt{1-p^2(v_0+az)^2}}-p^22(v_0+az)adz=\frac{-p^2a(v_0+az)adz}{\sqrt{1-p^2(v_0+az)^2}}$$

Deste modo,

$$X=-\frac{1}{ap}\int du=-\frac{1}{ap}u\big|_{0}^{H}=-\frac{1}{ap}\sqrt{1-p^2(v_0+az)^2}\big|_{0}^{H}$$

Retomando,

$$X=-\frac{1}{ap}\sqrt{1-p^2(v_0+az)^2}\big|_{0}^{H}$$

$$X=-\frac{1}{ap}\sqrt{1-p^2(v_0+aH)^2}+\frac{1}{ap}\sqrt{1-p^2v^2_0}$$

Passando a segunda parcela do lado direito para o lado esquerdo subtraindo,

$$X-\frac{1}{ap}\sqrt{1-p^2v^2_0}=-\frac{1}{ap}\sqrt{1-p^2(v_0+aH)^2}$$

Elevando os dois lados da expressão ao quadrado,

$$\left[X-\frac{1}{ap}\sqrt{1-p^2v^2_0}\right]^2=\frac{1}{a^2p^2}(1-p^2(v_0+aH)^2)$$

O lado direito da expressão,

$$\frac{1}{a^2p^2}(1-p^2(v_0+aH)^2)=\frac{1}{a^2p^2}-\frac{1}{a^2}(v_0+aH)^2=\frac{1}{a^2p^2}-\frac{a^2}{a^2}\left(\frac{v_0}{a}+H\right)^2=\frac{1}{a^2p^2}-\left(\frac{v_0}{a}+H\right)^2$$

Pois,

$$(v_0+aH)^2=\left[a\left(\frac{v_0}{a}+H\right)\right]^2=a^2\left(\frac{v_0}{a}+H\right)^2$$

Deste modo,

$$\left[X-\frac{1}{ap}\sqrt{1-p^2v^2_0}\right]^2=\frac{1}{a^2p^2}-\left[\frac{v_0}{a}+H\right]^2$$

Passando a segunda parcela do lado direito para o lado esquerdo subtraindo,

$$\left[X-\frac{1}{ap}\sqrt{1-p^2v^2_0}\right]^2+\left[\frac{v_0}{a}+H\right]^2=\frac{1}{a^2p^2}$$

Esta é a equação de um arco circular de raio $\frac{1}{ap}$, onde,

$$(X_C,Z_C)=\left(\frac{\sqrt{1-p^2v^2_0}}{ap},-\frac{v_0}{a}\right)$$
