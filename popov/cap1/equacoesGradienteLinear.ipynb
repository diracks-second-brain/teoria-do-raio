# Variação linear da velocidade com a profundidade

O modelo de em que a velocidade cresce linearmente com a profundidade é definido pela função:

$$v(z)=v_0+az$$

Onde $a$ é o gradiente linear de velocidade:

$$a\geq0$$

Para este modelo, no _turning point_,

$$pv(z)=p(v_0+az)=1$$

Então,

$$z_{tp}=\frac{1-pv_0}{pa}$$

Partindo da equação do tempo de trânsito $T$:

$$T=2\int_{0}^{H}\frac{dz}{v(z)\sqrt{1-p^2v^2(z)}}$$

Se,

$$y=p(v_0+az)$$

$$dy=padz$$

$$y(0)=pv_0$$

$$y(H)=pv_0+paH=pv_H$$

A integral para T,

$$T=\frac{2}{a}\int_{0}^{H}\frac{apdz}{pv(z)\sqrt{1-p^2v^2(z)}}=\frac{2}{a}\int_{0}^{H}\frac{dy}{y\sqrt{1-y^2}}$$

O resultado da integral é definido na tabela de integrais. Assim,

$$\int_{0}^{H}\frac{dy}{y\sqrt{1-y^2}}=-\ln\Bigg|\frac{1+\sqrt{1^2-y^2}}{y}\Bigg|$$

Deste modo,

$$T=\frac{-2}{a}\ln\Bigg|\frac{1+\sqrt{1-y^2}}{y}\Bigg|_{y=pv_0}^{y=pv_H}$$

Invertendo os limites,

$$T=\frac{2}{a}\ln\Bigg|\frac{1+\sqrt{1-y^2}}{y}\Bigg|_{y=pv_H}^{y=pv_0}$$

Logo,

$$T=\frac{2}{a}\left\{\ln\Bigg|\frac{1+\sqrt{1-p^2v^2_0}}{pv_0}\Bigg|-\ln\Bigg|\frac{1+\sqrt{1-p^2v^2_H}}{pv_H}\Bigg|\right\}$$

Utilizando a propriedade dos logarítmos,

$$T=\frac{2}{a}\ln\left[\frac{1+\sqrt{1-p^2v^2_0}}{pv_0}\frac{pv_H}{1+\sqrt{1-p^2v^2_H}}\right]$$

Por fim,

$$T=\frac{2}{a}\ln\left[\frac{v_H}{v_0}\frac{1+\sqrt{1-p^2v^2_0}}{1+\sqrt{1-p^2v^2_H}}\right]$$
