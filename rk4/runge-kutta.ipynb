# RUNGE-KUTTA DE QUARTA ORDEM 2D

Nós utilizamos o método Runge-Kutta de quarta ordem 2D para obter a solução dos sistemas dinâmico e cinemático do raio. O método Runge-Kutta
é um método de intergração numérica desenvolvido por Carl David Tolmé Runge em 1895, seguido de uma solução geral em 1901, elaborada por Martin Wilhelm Kutta \cite{algo}.
Para o seguinte sistema de equações diferenciais 2D:

\begin{equation}
\label{eq:ap:2.0.1}
 x'=f(t,x,y)
\end{equation}

\begin{equation}
\label{eq:ap:2.0.1}
 y'=g(t,x,y)
\end{equation}

A solução do sistema, as funções $x(t)$ e $y(t)$, é obtida de maneira iterativa a partir do cálculo das seguintes inclinações $k$'s e $l$'s.

\begin{equation}
\label{eq:ap:2.1}
k_1=f(t_n,x_n,y_n)
\end{equation}

\begin{equation}
\label{eq:ap:2.2}
l_1=g(t_n,x_n,y_n)
\end{equation}

\begin{equation}
\label{eq:ap:2.3}
k_2=f(t_n+\frac{1}{2}h,x_n+\frac{1}{2}hk_1,y_n+\frac{1}{2}hl_1)
\end{equation}

\begin{equation}
\label{eq:ap:2.4}
l_2=g(t_n+\frac{1}{2}h,x_n+\frac{1}{2}hk_1,y_n+\frac{1}{2}hl_1)
\end{equation}

\begin{equation}
\label{eq:ap:2.5}
k_3=f(t_n+\frac{1}{2}h,x_n+\frac{1}{2}hk_2,y_n+\frac{1}{2}hl_2)
\end{equation}

\begin{equation}
\label{eq:ap:2.6}
l_3=g(t_n+\frac{1}{2}h,x_n+\frac{1}{2}hk_2,y_n+\frac{1}{2}hl_2)
\end{equation}

\begin{equation}
\label{eq:ap:2.7}
k_4=f(t_n+h,x_n+hk_3,y_n+hl_3)
\end{equation}

\begin{equation}
\label{eq:ap:2.8}
l_4=g(t_n+h,x_n+hk_3,y_n+hl_3)
\end{equation}

A partir das inclinações, $k$'s e $l$'s, estima-se a inclinação $k$ e a inclinação $l$:

\begin{equation}
\label{eq:ap:2.9}
k=\frac{1}{6}(k_1+2k_2+2k_3+k_4)
\end{equation}

\begin{equation}
\label{eq:ap:2.10}
l=\frac{1}{6}(l_1+2l_2+2l_3+l_4)
\end{equation}

E deste modo, os valores de $t$, $x$ e $y$ são atualizados:

\begin{equation}
\label{eq:ap:2.11}
t_{n+1}=t_n+h
\end{equation}

\begin{equation}
\label{eq:ap:2.12}
x_{n+1}=x_n+hk
\end{equation}

\begin{equation}
\label{eq:ap:2.13}
y_{n+1}=y_n+hl
\end{equation}

O processo se repete para $N$ pontos das funções $x$ e $y$.
